```
#### importing the essential libraries
import matplotlib.pyplot as plt
import numpy as np
import cv2 
```

```
#reading the image
img = cv2.imread('image1.jpg', cv2.IMREAD_GRAYSCALE)
```

```
#showing the image
plt.imshow(img,cmap = 'gray')
```

```
im2 = cv2.imread('image2.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(im2,cmap = 'gray')
```

```
#cropping the image
img3 = img[400:733,200:700]
print(img3.shape)
plt.imshow(img3, cmap = 'gray')
```

```
#adding the two images 
#the bright area gets brighter
im4 = img3 + im2
```

```
plt.imshow(im4,cmap = 'gray')
```

```
#guassian distribution
image3 = np.random.randint(333,500)
print(image3)
```

```
plt.imshow(im2+image3,cmap = 'gray')
```
